```python
'''
W7A2
Cho danh sách arr và một số nguyên x, đếm xem x 
xuất hiện bao nhiêu lần trong danh sách. def count_occurrences(arr, x) -> int
'''
def count_occurrences(arr,x):
    count=0
    for i in arr :
        if i==x:
            count+=1
    return count
arr=list(map(int,input().split()))
x=int(input())
print(count_occurrences(arr,x))
```

```python
# W7A2 solution 2 :
def count_occurences(arr,x):
    return arr.count(x)
arr=list(map(int,input().split()))
x=int(input())
print(count_occurrences(arr,x))
```

```python
'''
W7A3
Viết hàm thực hiện thuật toán bubble sort, Selection 
Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp 
xếp nổi bọt (Bubble Sort);
'''
def bubble_sort(arr): 
    for i in range(len(arr)):
        for j in range(0,len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j+1],arr[j]=arr[j],arr[j+1]
    return arr
arr=list(map(int,input().split()))
print(bubble_sort(arr))
            
        
```

```python
# W7A3 selection_sort chọn min  đổi chỗ cho vị trí đầu tiên  
def selection_sort(arr): # Tìm min để đổi
    n=len(arr)
    for i in range(n):
        min_idx= i
        for j in range(i+1,n):
            if arr[min_idx] > arr[j]:
                min_idx= j
        # đổi chỗ sau khi tìm xong min 
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
        
arr=list(map(int,input().split()))
print(selection_sort(arr))
```

```python
'''
W7A4
Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất 
trong danh sách. Nếu có nhiều phần tử cùng số lần, 
trả về phần tử xuất hiện trước
'''
def dic(arr):
    dic={}
    for i in arr:
            if i in dic:
                dic[i]+=1
            else:
                dic[i]=1
    return dic
arr=list(map(int,input().split()))
d = dic(arr)
max_value=max(d.values())
my_list=[]
for key in d.keys():
    if d[key] == max_value:
        my_list.append(key)
print(d)
print(my_list[0])
    
 
```
