```python
'''
Bài tập xác định ngày tháng năm hợp lệ
'''
from datetime import date
ngay= int(input("Nhập ngày  :"))
thang= int(input("Nhập tháng  :"))
nam= int(input("Nhập năm :"))
if ngay<0 or ngay>31 or thang<0 or thang>12 or nam>1900 or nam<2100 :
   print("Không hợp lệ")
elif thang==2 and (nam%4==0 and nam%100!=0) or (nam%400==0) :
   if ngay<=29:
             print("Hợp lệ")
   else:
             print("Không hợp lệ")
elif thang==4 or thang==6 or thang==9 or thang==11:
    if ngay<=30:
      print("Hợp lệ")
    else:
      print("Không hợp lệ")
else:
    print("Hợp lệ")
```
