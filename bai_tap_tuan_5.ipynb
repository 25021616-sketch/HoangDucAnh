```python
'''
W5A1
Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two(
a, b) -> int:
'''
def Max_Of_Two(a, b):
  if a>b:
    return a
  else:
    return b
a=int(input("nhập số a "))
b=int(input("nhập số b "))
print(Max_Of_Two(a, b))
```

```python
'''
W5A2 Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]:
'''
def swap(a:int,b:int):
  c=a+b
  b=c-b
  a=c-a
  return a,b
a=int(input("nhập số a "))
b=int(input("nhập số b "))
a,b=swap(a,b)
print(f"số a sau khi hoán đổi là {a}")
print(f"số b sau khi hoán đổi là {b}")
```

```python
'''
W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
'''
def so_nguyen_to(n):
  if n<2:
    return False
  else:
    for i in range(2,int(n**0.5)+1):
      if n%i==0:
        return False
    return True
n=int(input("nhập số n "))
if so_nguyen_to(n):
  print(f"{n} là số nguyên tố")
else:
  print("không phải số nguyên tố")
```

```python
def matrix(m,n):
  matrix=[]
  for i in range(m):
    matrix_=list(map(int,input(f"Nhập {n} phần tử ở dòng {i+1}:").split()))
    if len(matrix_)==n:
      matrix.append(matrix_)
    else:
      return None
  return matrix
m=int(input("nhập số dòng "))
n=int(input("nhập số cột "))
matrix_A=matrix(m,n)
if matrix_A:
  for r in matrix_A:
    print(r)
```

```python
'''
W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không,
xử lý trên dữ liệu kiểu số nguyên
'''
def  perfect_number(n):
  total=0
  if n<=0:
    return False
  else:
    for i in range(1,int(n/2)+1):
      if n%i==0:
        total+=i
    return total==n
n=int(input("Enter a positive integer n :"))
if perfect_number(n):
   print(f"{n} is perfect_number")
else:
   print(f"{n} is't perfect_number")
```

```python
'''
W5A5
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của
phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1
'''
def find_first_index(number,k):
    for i in range(len(number)):
        if number[i]==k:
            return i+1
    return -1
number=list(map(int,input("Input a list of positive integers ").split()))
k=int(input("Enter a number to find:"))
print (find_first_index(number,k))
```

```python
'''
W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)
'''
def factorial(n):# giai_thừa
  result=1 #result: kết quả tạm thời
  if n<0:
    return None
  else:
    for i in range(1,n+1):
      result*=i
    return result
n=int(input("enter a positive integer "))
print(factorial(n))
```

```python
'''
W5A7

'''
def toan_tu(num1,num2,operat):
    if operat=='+':
        result=num1+num2
    elif operat=='-':
        result=num1-num2
    elif operat=='*':
        result=num1*num2
    elif operat=='/' and num2!=0:
        result=num1/num2
    else:
        return 'Error: Invalid operator'
    return round(result,2)
num1=float(input("enter a number 1:"))
num2=float(input("enter a number 2:"))
operat=input("Chọn{+};{-};{*};{/}:")
print(toan_tu(num1,num2,operat))
```
