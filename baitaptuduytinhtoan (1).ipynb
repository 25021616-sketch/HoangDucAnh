```python
'''
Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân.
'''
a=int(input("nhập chiều rộng hình chữ nhật :"))
b=int(input("nhập chiều dài hình chữ nhật :"))
s=a*b-(a/2)**2*3.14
if a>b:
  print("đây không phải hình chữ nhật")
else:
  print(f"diện tích còn lại trồng cây là :{s:.2f}")
```

```python
'''
Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.
'''
import math
a,b,c= map(float,input('Nhập lần lượt độ dài 3 cạnh của tam giác từ lớn đến bé :').split())
d= (a+b+c)/2
if a+b<=c or a+c<=b or b+c<=a:
    print("Đây không phải là 1 tam giác")
else:
    print(f'Diện tích tam giác là:{math.sqrt(d*(d-a)*(d-b)*(d-c)):.1f}')
```

```python
'''
Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng
'''
ten = input("Nhập họ tên chủ hộ: ")
chi_so_cu = int(input("Nhập chỉ số điện kế tháng trước: "))
chi_so_moi = int(input("Nhập chỉ số điện kế tháng này: "))

# Số kWh đã dùng
so_kwh = chi_so_moi - chi_so_cu
tien_dien = 0

# Tính tiền điện theo bậc thang
if so_kwh <= 50:
    tien_dien = so_kwh * 1984
elif so_kwh <= 100:
    tien_dien = 50 * 1984 + (so_kwh - 50) * 2050
elif so_kwh <= 200:
    tien_dien = 50 * 1984 + 50 * 2050 + (so_kwh - 100) * 2380
elif so_kwh <= 300:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (so_kwh - 200) * 2998
elif so_kwh <= 400:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (so_kwh - 300) * 3350
else:
    tien_dien = (50 * 1984 + 50 * 2050 + 100 * 2380 +
                 100 * 2998 + 100 * 3350 + (so_kwh - 400) * 3460)


tien_dien_vat = int(tien_dien * 1.08)   # chỉ lấy đến đơn vị đồng

# Xuất kết quả
print(f"\nChủ hộ: {ten}")
print(f"Số điện tiêu thụ: {so_kwh} kWh")
print(f"Số tiền phải trả (đã gồm VAT 8%): {tien_dien_vat} đồng")
```
