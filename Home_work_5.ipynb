```python
'''
W6A18
Nhâp vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k)
'''
def matrix(m,n):
  matrix_=[]
  for i in range(m):
    row=list(map(int,input().split()))
    if len(row)==n:
      matrix_.append(row)
    else:
      return None
  return matrix_
m,n,k=map(int,input().split())
matrix_A=matrix(m,n)
if matrix_A:
  for row in matrix_A:
    for x in row:
      print(f"{x:4d}",end=" ")
    print()
total_row=0
for i in range(m):
  total_row+=matrix_A[i][k]
print(total_row)
total_cloumn=0
for j in range(n):
  total_cloumn+=matrix_A[k][j]
print(total_cloumn)

```

```python
'''
W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1
'''
def find_list_index(n,k):
  for m in n:
    if m==k:
      return n.index(m)+1
  return -1
n=list(map(int,input().split()))
k=int(input())
a=find_list_index(n,k)
print(a)
```

```python
'''
W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9
'''
def push_0_to_front(n):
  zeros=[x for x in n if x==0]
  number=[x for x in n if x!=0]
  return zeros + number
def push_0_to_end(n):
  zeros=[x for x in n if x==0]
  number=[x for x in n if x!=0]
  return number + zeros
n=list(map(int,input().split()))
print(push_0_to_front(n))
print(push_0_to_end(n))
```

```python
'''
W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối
'''
my_list=list(map(int,input().split()))
result=[]
for i in range(0,len(my_list)-2):
  if my_list[i]<my_list[i+1]>my_list[i+2]:
    result.append(my_list[i+1])
print(sum(result))
```

```python
'''
W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1
'''
n=list(map(int,input().split()))
x=int(input())
result=[]
for i in range(len(n)):
  if n[i]<x:
    result.append(i)
if len(result)==0:
  print('-1')
else:
  print(result)
```

```python
'''
W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)
'''
n=list(map(int,input().split()))
is_increasing = True
is_decreasing = True
for i in range(len(n)-1):
  if n[i]>=n[i+1]:
    is_increasing = False
  if n[i]<=n[i+1]:
    is_decreasing= False
if is_increasing:
  print("Tang dan")
elif is_decreasing:
  print("giam dan")
else:
  print("vo danh")
```
