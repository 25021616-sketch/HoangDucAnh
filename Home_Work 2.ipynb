```python
'''
Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diệntíchvàchu vi của nó.
'''
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))

dien_tich = a * b
chu_vi = 2 * (a + b)

print("Diện tích hình chữ nhật là:", dien_tich)
print("Chu vi hình chữ nhật là:", chu_vi)
```

```python
'''
Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.
'''
import math

r = float(input("Nhập bán kính hình tròn: "))

dien_tich = math.pi * r**2
chu_vi = 2 * math.pi * r

print(f"Diện tích hình tròn là:{dien_tich:.2f} ")
print(f"Chu vi hình tròn là:{chu_vi:.2f}")













```

```python
'''
Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là
độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh
cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân,
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn
hình.
'''
import math
a,b,c= map(float,input('Nhập lần lượt độ dài 3 cạnh của tam giác  :').split())

if a+b<=c or a+c<=b or b+c<=a:
    print("Đây không phải là 1 tam giác")
if a+b>c and a+c>b and b+c>a:
    if a==b==c:
      print("Đây là tâm giác đều")
    elif a==b or a==c or b==c:
      print("Đây là tam giác cân")
    elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==b**2+a**2:
      print("Đây là tam giác vuông")
    else:
      print("Đây là tam giác thường")
    d= (a+b+c)/2
    print(f'Diện tích tam giác là:{math.sqrt(d*(d-a)*(d-b)*(d-c)):.2f}')
    print(f'Chu vi tam giác là:{a+b+c}')
```

```python
'''
Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)
'''
a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))

if a==0:
    if b==0 and c==0:
        print(f"Phương trình có vô số nghiệm")
    elif b==0 and c!=0:
        print("Phương trình vô nghiệm")
    else:
        print(f"Nghiệm của phương trình là x = {-c/b}")
else:
    detal=b**2-4*a*c
    if detal < 0:
        print("Phương trình vô nghiệm" )
    elif detal==0:
        print(f"Phương trình có nghiệm kép x = {-b/(2*a)}")
    else:
        print(f"Phương trình có 2 nghiệm lần lượt là : {(-b+ float(detal**0.5))/(2*a)} và {(-b-float(detal**0.5))/(2*a)} ")

```

```python
'''
Bài 8: Giải hệ phương trình tuyến tính:
ax  by  m
cx  dy  n
'''
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
m = float(input("Nhập m: "))
n = float(input("Nhập n: "))

# Định thức()
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c

if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ có nghiệm duy nhất: x = {x}, y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ có vô số nghiệm")
    else:
        print("Hệ vô nghiệm")
```

```python
'''
Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao
nhiêu giây
'''
x = int(input("Nhập thời gian (giây): "))

gio=x//3600
phut=(x%3600)//60
giay=x%60
print(f"{x} giây = {gio} giờ {phut} phút {giay} giây")
```
