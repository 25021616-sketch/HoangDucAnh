```python
'''
W7A16
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử
x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp 
hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn 
lại thành mảng đẹp
'''
def mang_dep(arr):
    dic={}
    for i in arr:
        if i in dic:
            dic[i]+=1
        else:
            dic[i]=1
    remove=0
    for keys,values in dic.items():
        if keys > values :
            remove+=values
        elif keys < values:
            remove+= values-keys
    return remove
arr=list(map(int,input().strip().split()))
if mang_dep(arr)==0:
    print("YES")
else:
    print(f"NO {mang_dep(arr)}")
            
            
        
    
    
                
    
```

```python
'''
W7A18
Sắp xếp có điều kiện: Số chẵn đứng 
trước, số lẻ đứng sau, chẵn tăng dần, lẻ
giảm dần
'''
def my_list(arr):
    so_chan=[]
    so_le=[]
    for i in arr:
        if i % 2 ==0:
            so_chan.append(i)
        else:
            so_le.append(i)
    so_chan.sort()
    so_le.sort(reverse=True)

    New_list= so_chan + so_le
    return New_list
arr=list(map(int,input().strip().split()))
print(my_list(arr))
```

```python
'''
W7A28
Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên tố nào <= N (N 
là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ 
nhất (gặp đầu tiên từ trái qua)
'''
n=int(input())
is_prime=[True]*(n+1)
is_prime[0]=is_prime[1]=False
result=[]
for i in range(2,int(n**0.5)+1):
    if is_prime[i]:
        for j in range(i*i,n+1,i):
            is_prime[j]=False
result=[ i for i in range(2,n+1) if is_prime[i]]
danh_sach=[x for x in range(1,n+1) if x not in result]
print(result)
print(danh_sach)
my_list=[]
tmp=[danh_sach[0]]
for i in range(1,len(danh_sach)):
    if danh_sach[i]==danh_sach[i-1]+1:
        tmp.append(danh_sach[i])
    else:
        my_list.append(tmp)
        tmp=[danh_sach[i]]
if tmp:
    my_list.append(tmp)
print(my_list)
max_len=max(len(x) for x in my_list)
max_len_list=[]
for i in my_list:
    if len(i)==max_len:
        max_len_list.append(i)
print(max_len_list[0])

    
    


    
```
